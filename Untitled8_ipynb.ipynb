```python
d = {'a': 30, 'b': 10, 'c': 20}

print("Original Dictionary:", d)

# 1. Ascending Order
asc = sorted(d.items(), key=lambda x: x[1])
print("Ascending Order:", asc)

#  Descending Order
desc = sorted(d.items(), key=lambda x: x[1], reverse=True)
print("Descending Order:", desc)

# 2. Check key exists
if 'a' in d:
    print("Key exists")
else:
    print("Key does not exist")

# 3. Merge dictionary
d2 = {'e': 50}

d.update(d2)

print("After Merging:", d)

# 4. Add item
d['d'] = 40
print("After Adding Item:", d)

# 5.Create a tuple with different data types
d = {'a': 30, 'b': 10, 'c': 20}

t = ('a', 30, 'b', 10, 'c', 20)

print("Dictionary:", d)
print("Tuple:", t)

# 6.Sum All Items in List
numbers = [10, 20, 30, 40, 50]

total = sum(numbers)

print("Sum =", total)

# 7.Largest number in List
numbers = [10, 20, 30, 40, 50]

largest = max(numbers)

print("Largest Number =", largest)

# 8.Add members in set
s = {10, 20, 30, 40, 50}

s.add(60)

print("Updated Set:", s)

# 9.Reverse array items
from array import *
arr = array('i', [10, 20, 30, 40, 50])

arr.reverse()

print("Reversed Array:", arr)

# 10.Create Array and Access Elements Using Index
from array import *

arr = array('i', [10, 20, 30, 40, 50])

for i in range(5):
    print("Element at index", i, "=", arr[i])
```
