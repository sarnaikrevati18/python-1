```python

for i in range(1, 11):
    if i % 2 == 0:
        print(i)
```

```python

sum_odd = 0

for i in range(1, 11):
    if i % 2 != 0:
        sum_odd += i

print("Sum of odd numbers:", sum_odd)
```

```python

a, b = 0, 1

print("Fibonacci series:")
while a <= 50:
    print(a, end=" ")
    a, b = b, a + b
```

```python
#
string = input("Enter a string: ")

result = ""
for i in range(len(string)):
    if i % 2 == 0:
        result += string[i]

print("Result:", result)
```
