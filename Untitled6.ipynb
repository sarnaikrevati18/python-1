```python
# Create a list
numbers = [10, 20, 30, 40, 50]

# Check membership
print(20 in numbers)
print(60 not in numbers)
```

```python
# Create a list
fruits = ["apple", "banana", "mango", "grape", "orange"]

# Indexing
print(fruits[0])
print(fruits[2])

# Negative indexing
print(fruits[-1])
print(fruits[-3])

# Slicing
print(fruits[1:4])
print(fruits[:3])
print(fruits[2:])
```

```python
# Create a list
nums = [1, 2, 3, 4]

# Update
nums[1] = 20
print(nums)

# Append
nums.append(5)
print(nums)

# Insert
nums.insert(2, 15)
print(nums)

# Delete
nums.remove(3)
print(nums)

del nums[0]
print(nums)
```

```python
# Create lists
list1 = [1, 2, 3]
list2 = [4, 5]

# Concatenation
print(list1 + list2)

# Repetition
print(list1 * 2)

# Length
print(len(list1))

# Maximum
print(max(list1))

# Minimum
print(min(list1))
```
